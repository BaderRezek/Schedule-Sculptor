# Catalog helpers from Python modules
This notebook shows how to pull a function out of `rag/data/raw/catalog.py` so
you can reuse it while exploring data.

```python
from pathlib import Path
import sys

project_root = None
for candidate in (Path.cwd().resolve(), *Path.cwd().resolve().parents):
    if (candidate / "rag").is_dir():
        project_root = candidate
        break

if project_root is None:
    raise RuntimeError("Could not locate the project root containing the 'rag' package.")

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from rag.data.raw.catalog import normalize_course_title

normalize_course_title("   introduction TO   ai  ")
```
